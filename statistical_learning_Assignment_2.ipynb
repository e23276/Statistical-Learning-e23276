# Q3) Transformer Failures: Formal Analysis

### 1. Probability Space $(\Omega, \mathcal{F}, P)$
* **Sample Space ($\Omega$):** The set of all possible outcomes.
  $$\Omega = \{0, 1, 2, 3\}$$
* **$\sigma$-algebra ($\mathcal{F}$):** The collection of all possible events. For this finite case:
  $$\mathcal{F} = \mathcal{P}(\Omega)$$  
  #### using power set we can obtain the 16 algebra
* **Probability Measure ($P$):** Defined for each outcome:
  $$P(\{0\}) = 0.50, \quad P(\{1\}) = 0.30, \quad P(\{2\}) = 0.15, \quad P(\{3\}) = 0.05$$

### 2. Random Variable Definition
$X$ is defined as the identity function on $\Omega$, mapping each outcome to its numerical value:
$$X: \Omega \to \{0, 1, 2, 3\}, \quad X(\omega) = \omega$$

### 3. Probability Mass Function (PMF)
The PMF $p_X(x) = P(X = x)$ is:
| $x$ | $P(X = x)$ |
|---|---|
| 0 | 0.50 |
| 1 | 0.30 |
| 2 | 0.15 |
| 3 | 0.05 |

### 4. Expectation $E[X]$
$$E[X] = \sum x \cdot P(X=x) = (0 \cdot 0.5) + (1 \cdot 0.3) + (2 \cdot 0.15) + (3 \cdot 0.05) = \mathbf{0.75}$$

### 5. Variance $Var(X)$
First, compute $E[X^2]$:
$$E[X^2] = (0^2 \cdot 0.5) + (1^2 \cdot 0.3) + (2^2 \cdot 0.15) + (3^2 \cdot 0.05) = 1.35$$
$$Var(X) = E[X^2] - (E[X])^2 = 1.35 - (0.75)^2 = \mathbf{0.7875}$$

### 6. Probability Ranges
* **$P(X \geq 1)$:** $1 - P(X=0) = 1 - 0.50 = \mathbf{0.50}$
* **$P(X \geq 2)$:** $P(X=2) + P(X=3) = 0.15 + 0.05 = \mathbf{0.20}$

### 7. Practical Meaning
* **Expectation:** The predicted average number of failures per day over a long period.
* **Variance:** The degree of daily fluctuation/risk around that average.

# Q4) Probability of Defective Items: Solutions

### 1. Construct a probability space $(\Omega, \mathcal{F}, P)$
To model the proportion of defective items $X \in [0, 1]$:
* **Sample Space ($\Omega$):** The set of all possible proportions, $\Omega = [0, 1]$.
* **$\sigma$-algebra ($\mathcal{F}$):** The Borel $\sigma$-algebra on $[0, 1]$, denoted as $\mathcal{B}([0, 1])$.
* **Probability Measure ($P$):** Defined by the Beta distribution density, where for any event $A \in \mathcal{F}$:
    $$P(A) = \int_A f(x) \, dx$$

### 2. Define $X$ as a continuous random variable
$X$ is the identity mapping $X: \Omega \to [0, 1]$ such that $X(\omega) = \omega$. Since the mapping is measurable and the range is continuous, $X$ is a continuous random variable.

### 3. Probability Density Function (PDF)
Given $\alpha = 2$ and $\beta = 18$, the PDF is:
$$f(x) = \begin{cases} \frac{1}{B(2, 18)} x^{2-1} (1-x)^{18-1}, & 0 < x < 1 \\ 0, & \text{otherwise} \end{cases}$$
Simplifying the exponents:
$$f(x) = \frac{1}{B(2, 18)} x(1-x)^{17}$$

### 4. Verify that $f$ is a valid PDF
A valid PDF must satisfy $\int_{-\infty}^{\infty} f(x) \, dx = 1$.
Using the definition of the Beta function $B(\alpha, \beta) = \int_0^1 x^{\alpha-1}(1-x)^{\beta-1} \, dx$:
$$\int_0^1 \frac{1}{B(2, 18)} x^{2-1}(1-x)^{18-1} \, dx = \frac{1}{B(2, 18)} \int_0^1 x^{2-1}(1-x)^{18-1} \, dx$$
$$\int_0^1 f(x) \, dx = \frac{B(2, 18)}{B(2, 18)} = 1$$

### 5. Compute the Expectation $E[X]$
For a Beta distribution, $E[X] = \frac{\alpha}{\alpha + \beta}$:
$$E[X] = \frac{2}{2 + 18} = \frac{2}{20} = \mathbf{0.10}$$

### 6. Compute the Variance $Var(X)$
For a Beta distribution, $Var(X) = \frac{\alpha\beta}{(\alpha + \beta)^2(\alpha + \beta + 1)}$:
$$Var(X) = \frac{2 \times 18}{(20)^2(20 + 1)} = \frac{36}{400 \times 21} = \frac{36}{8400} = \mathbf{0.004286}$$

### 7. Probability $P(X \leq 0.10)$
In Beta integral form:
$$P(X \leq 0.10) = \frac{1}{B(2, 18)} \int_0^{0.10} x(1-x)^{17} \, dx$$

### 8. Probability $P(0.05 \leq X \leq 0.15)$
In integral form:
$$P(0.05 \leq X \leq 0.15) = \frac{1}{B(2, 18)} \int_{0.05}^{0.15} x(1-x)^{17} \, dx$$

### 9. Why is Beta distribution a reasonable model?
The Beta distribution is ideal because its domain is strictly bounded between **0 and 1**, which matches the physical constraints of a "proportion" or "percentage." Additionally, by adjusting $\alpha$ and $\beta$, it can model various shapes of uncertainty for probabilities.

### 10. Interpretation
* **Expectation (0.10):** On average, the manufacturing process produces 10% defective items. It represents the "center of gravity" of the defect rate.
* **Variance (0.0043):** This represents the consistency of the process. A low variance indicates that the daily defect rate stays very close to 10%, while a high variance would suggest the process is unstable and unpredictable.

# Q5) Expected Project Completion Time: Solutions

Given parameters for the Triangular distribution:
* Lower limit $a = 4$
* Upper limit $b = 10$
* Mode $c = 6$

---

### 1. Support of $X$
The support of the random variable $X$ is the interval of all possible values for the completion time:
$$\text{Support} = [a, b] = [4, 10]$$

### 2. Probability Density Function $f_X(x)$
For a triangular distribution, the PDF is defined piecewise:
$$f_X(x) = \begin{cases}
\frac{2(x-a)}{(b-a)(c-a)} & \text{for } a \leq x < c \\
\frac{2(b-x)}{(b-a)(b-c)} & \text{for } c \leq x \leq b \\
0 & \text{otherwise}
\end{cases}$$

Plugging in $a=4, b=10, c=6$:
$$f_X(x) = \begin{cases}
\frac{2(x-4)}{(6)(2)} = \frac{x-4}{6} & \text{for } 4 \leq x < 6 \\
\frac{2(10-x)}{(6)(4)} = \frac{10-x}{12} & \text{for } 6 \leq x \leq 10
\end{cases}$$

### 3. Verification
To verify it integrates to 1, we calculate the area of the triangle:
$$\text{Area} = \frac{1}{2} \times \text{base} \times \text{height} = \frac{1}{2} \times (b-a) \times \frac{2}{b-a} = \frac{1}{2} \times 6 \times \frac{2}{6} = 1$$

### 4. Expectation $E[X]$
$$E[X] = \frac{a + b + c}{3} = \frac{4 + 10 + 6}{3} = \frac{20}{3} \approx \mathbf{6.67 \text{ hours}}$$

### 5. Variance $Var(X)$
$$Var(X) = \frac{a^2 + b^2 + c^2 - ab - ac - bc}{18} = \frac{4^2 + 10^2 + 6^2 - (40) - (24) - (60)}{18}$$
$$Var(X) = \frac{16 + 100 + 36 - 124}{18} = \frac{28}{18} \approx \mathbf{1.56}$$

### 6. Interpretation of $a, b,$ and $c$
* **$a = 4$:** The "best-case scenario" (shortest possible time).
* **$b = 10$:** The "worst-case scenario" (longest possible time).
* **$c = 6$:** The "most likely" time (peak of the distribution).

### 7. Why use a Triangular Distribution?
It is a reasonable model because it only requires three intuitive data points (minimum, maximum, and mode), which are easily provided by expert judgment when historical data is scarce.

### 8. Probability $P(X \leq 5)$
Since $5 < c$, we use the first part of the CDF:
$$P(X \leq x) = \frac{(x-a)^2}{(b-a)(c-a)} \implies P(X \leq 5) = \frac{(5-4)^2}{(6)(2)} = \frac{1}{12} \approx \mathbf{0.0833}$$

### 9. Probability $P(5 \leq X \leq 8)$
$P(5 \leq X \leq 8) = P(X \leq 8) - P(X \leq 5)$.
For $x=8$ (where $x > c$):
$$P(X \leq 8) = 1 - \frac{(b-x)^2}{(b-a)(b-c)} = 1 - \frac{(10-8)^2}{(6)(4)} = 1 - \frac{4}{24} = \frac{5}{6}$$
$$P(5 \leq X \leq 8) = \frac{5}{6} - \frac{1}{12} = \frac{10-1}{12} = \frac{9}{12} = \mathbf{0.75}$$

### 10. Project Planning and Sensitivity Analysis
* **Planning:** The expectation (6.67h) helps in scheduling labor, while the probabilities help estimate the risk of delays.
* **Sensitivity:** Managers can change $b$ (e.g., if a part delivery is late) to see how much the average completion time shifts.

# Q6) Expected Project Completion Time Using PERT: Solutions

Given parameters:
* Minimum time ($a$) = 8 days
* Maximum time ($b$) = 20 days
* Most likely time ($c$) = 11 days

### 1. Compute shape parameters $\alpha$ and $\beta$
Using the provided formulas:
$$\alpha = 1 + 4\left(\frac{c - a}{b - a}\right) = 1 + 4\left(\frac{11 - 8}{20 - 8}\right) = 1 + 4\left(\frac{3}{12}\right) = 1 + 1 = \mathbf{2}$$
$$\beta = 1 + 4\left(\frac{b - c}{b - a}\right) = 1 + 4\left(\frac{20 - 11}{20 - 8}\right) = 1 + 4\left(\frac{9}{12}\right) = 1 + 3 = \mathbf{4}$$

### 2. Specify the support of $X$
The support is the range of possible days for task completion:
$$\text{Support} = [a, b] = [8, 20]$$

### 3. Probability Density Function $f_X(x)$
Using $\alpha=2, \beta=4, a=8, b=20$:
$$f_X(x) = \frac{(x - 8)^{2-1}(20 - x)^{4-1}}{B(2, 4)(20 - 8)^{2+4-1}} = \frac{(x - 8)(20 - x)^3}{B(2, 4)(12)^5}, \quad 8 \leq x \leq 20$$

### 4. Probability Space $(\Omega, \mathcal{F}, P)$
* **Sample Space ($\Omega$):** $[8, 20]$
* **$\sigma$-algebra ($\mathcal{F}$):** Borel $\sigma$-algebra $\mathcal{B}([8, 20])$
* **Probability Measure ($P$):** $P(A) = \int_A f_X(x) \, dx$ for $A \in \mathcal{F}$

### 5. Define $X$ as a random variable
$X$ is the identity map $X: \Omega \to [8, 20]$ where $X(\omega) = \omega$.

### 6. Show the expectation $E[X]$
For a PERT distribution, $E[X] = \frac{a + 4c + b}{6}$:
$$E[X] = \frac{8 + 4(11) + 20}{6} = \frac{8 + 44 + 20}{6} = \frac{72}{6} = \mathbf{12 \text{ days}}$$

### 7. Compute the variance $Var(X)$
Using the approximation formula:
$$Var(X) = \frac{(E[X] - a)(b - E[X])}{7} = \frac{(12 - 8)(20 - 12)}{7} = \frac{4 \times 8}{7} = \frac{32}{7} \approx \mathbf{4.57}$$

### 8. Compute the standard deviation $\sigma$
$$\sigma = \sqrt{Var(X)} = \sqrt{4.5714} \approx \mathbf{2.14 \text{ days}}$$

### 9. PERT vs. Triangular Distribution
The PERT distribution is often more appropriate because it is **"smoother"** and gives less weight to the extreme values ($a$ and $b$) compared to the Triangular distribution. While the Triangular mean for these values would be $(8+20+11)/3 = 13$, PERT's mean (12) is more heavily influenced by the most likely value ($c$), reflecting a more realistic "expert" estimate.

### 10. Practical Interpretation
* **$a, b, c$:** Represent the range of uncertainty (best, worst, and typical cases).
* **$E[X] = 12$:** The most realistic target date to use for the project schedule.
* **$\sigma = 2.14$:** Measures the risk; a higher $\sigma$ indicates greater uncertainty in the task duration, signaling that the project manager should perhaps add a larger buffer for this task.

# Q7) If the PERT Expectation is to Match Our Intuition

Given parameters:
* Minimum time ($a$) = 5 days
* Most likely time ($c$) = 11 days
* Condition: Expected time $E[X]$ is equal to the most likely time $c$.

### 1. Determine the value of the maximum time $b$
The formula for the expected value (mean) of a PERT distribution is:
$$E[X] = \frac{a + 4c + b}{6}$$

According to the problem, we want $E[X] = c$. We set up the equation:
$$\frac{a + 4c + b}{6} = c$$

Now, solve for $b$:
$$a + 4c + b = 6c$$
$$b = 6c - 4c - a$$
$$b = 2c - a$$

Substitute the given values $a = 5$ and $c = 11$:
$$b = 2(11) - 5$$
$$b = 22 - 5$$
$$b = \mathbf{17 \text{ days}}$$

**Result:** For the expectation to match the most likely time, the maximum time $b$ must be 17 days.

---

### 2. What happens to the variance?
In a PERT distribution, the variance is typically approximated by:
$$Var(X) = \left( \frac{b - a}{6} \right)^2 \quad \text{or} \quad Var(X) = \frac{(E[X] - a)(b - E[X])}{7}$$

When $E[X] = c$, the distribution becomes **symmetric**.

* **Symmetry:** Since $b - c = 17 - 11 = 6$ and $c - a = 11 - 5 = 6$, the mode is exactly in the center of the range $[a, b]$.
* **Effect on Variance:** When the distribution is symmetric ($c$ is the midpoint), the variance is at its **maximum** for a fixed range $(b-a)$. If the distribution were skewed (if $c$ were closer to $a$ or $b$), the mass would be more concentrated, and the variance would decrease.
* **Numerical Calculation:**
  $$Var(X) = \frac{(11 - 5)(17 - 11)}{7} = \frac{6 \times 6}{7} = \frac{36}{7} \approx \mathbf{5.14}$$

**Conclusion:** When the expectation matches the most likely time in this specific scenario, the distribution is perfectly symmetric, and the variance reflects the spread of a balanced, bell-shaped Beta curve.
