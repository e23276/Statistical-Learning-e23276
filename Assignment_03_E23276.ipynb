##Q1) Prediction of Fraudulent Transactions
1. Probability Space $(\Omega, \mathcal{F}, P)$Sample Space ($\Omega$): The set of all possible outcomes is $\Omega = \{\omega_1, \omega_2, \omega_3, \omega_4\}$.  $\sigma$-algebra ($\mathcal{F}$): Since $\mathcal{F} = \mathfrak{P}(\Omega)$, it contains all $2^4 = 16$ possible subsets of $\Omega$.  Probability Measure ($P$): Defined by the given weights:$P(\{\omega_1\}) = 0.50$   $P(\{\omega_2\}) = 0.10$   $P(\{\omega_3\}) = 0.20$   $P(\{\omega_4\}) = 0.20$  


2. $\sigma$-algebra generated by $Y$, $\sigma(Y)$The random variable $Y$ groups the sample space into risk classes:$\{Y=0\} = \{\omega_1, \omega_2\}$ (Low-risk).  $\{Y=1\} = \{\omega_3, \omega_4\}$ (High-risk).  The $\sigma$-algebra generated by $Y$ consists of all unions of these sets:$$\sigma(Y) = \{ \emptyset, \Omega, \{\omega_1, \omega_2\}, \{\omega_3, \omega_4\} \}$$


3. Measurability of $E[X|Y]$By definition, the conditional expectation $E[X|Y]$ must be $\sigma(Y)$-measurable. This means that for the function to be valid, it must be constant over the "atoms" (the smallest non-empty sets) of the $\sigma$-algebra. Since the atoms are $\{\omega_1, \omega_2\}$ and $\{\omega_3, \omega_4\}$, the value of $E[X|Y]$ cannot change within those sets

4. Explicit Calculation of $E[X|Y]$We calculate the expected value of $X$ (fraud indicator) for each risk class:For $Y=0$: $E[X|Y=0] = \frac{P(X=1 \cap Y=0)}{P(Y=0)} = \frac{0.10}{0.50 + 0.10} = \frac{1}{6}$.  For $Y=1$: $E[X|Y=1] = \frac{P(X=1 \cap Y=1)}{P(Y=1)} = \frac{0.20}{0.20 + 0.20} = \frac{1}{2}$.  Result: $E[X|Y] = \frac{1}{6}\mathbb{1}_{\{Y=0\}} + \frac{1}{2}\mathbb{1}_{\{Y=1\}}$

5. Verify the Defining PropertyFor every set $A \in \sigma(Y)$, we must check if $\int_{A}E[X|Y]dP = \int_{A}XdP$.  Let $A = \{Y=0\}$:$\int_{A}E[X|Y]dP = \frac{1}{6} \times P(Y=0) = \frac{1}{6} \times 0.60 = 0.10$.$\int_{A}XdP = X(\omega_1)P(\omega_1) + X(\omega_2)P(\omega_2) = (0 \times 0.50) + (1 \times 0.10) = 0.10$.The values match, satisfying the property.

6. Conditional PMF $p_{X|Y}(x|y)$

7. Verify the IdentityThe identity $\mathbb{E}[X|Y=y] = \sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x|y)$ holds:  For $y=0$: $(0 \cdot 5/6) + (1 \cdot 1/6) = 1/6$.For $y=1$: $(0 \cdot 1/2) + (1 \cdot 1/2) = 1/2$.These match our values in Part 4.

8. Mean Square Error (MSE)The MSE is $E[(X - E[X|Y])^2]$:
$MSE = \sum_{\omega \in \Omega} (X(\omega) - E[X|Y](\omega))^2 P(\omega)$
$MSE = (0 - 1/6)^2(0.50) + (1 - 1/6)^2(0.10) + (0 - 1/2)^2(0.20) + (1 - 1/2)^2(0.20)$

Relationship: By the Law of Total Variance, $Var(X) = E[Var(X|Y)] + Var(E[X|Y])$. Knowledge of $Y$ reduces the total uncertainty ($0.21$) by the variance of the estimates ($Var(E[X|Y])$), leaving only the residual MSE ($0.1833$).

##Q2) Conditional Density of the Multivariate Gaussian
1. Derivation of the Conditional Distribution $f_{X|Y}(x|y)$To find the conditional density, we use the definition:$$f_{X|Y}(x|y) = \frac{f_{X,Y}(z)}{f_Y(y)}$$  The Joint DensityThe joint density of $Z = \begin{bmatrix} X \\ Y \end{bmatrix}$ is given by:$$f_{X,Y}(z) = \frac{1}{\sqrt{(2\pi)^{n+m}|\Sigma|}} \exp\left(-\frac{1}{2}(z-\mu)^{T}\Sigma^{-1}(z-\mu)\right)$$  Where the covariance matrix $\Sigma$ is partitioned as:$$\Sigma = \begin{bmatrix} \Sigma_{XX} & \Sigma_{XY} \\ \Sigma_{YX} & \Sigma_{YY} \end{bmatrix}$$  Completing the SquareBy using the properties of the Schur Complement and partitioning the inverse covariance matrix (the precision matrix), we can rewrite the quadratic form in the exponent. The term $(z-\mu)^{T}\Sigma^{-1}(z-\mu)$ is decomposed to isolate the terms involving $x$.The conditional distribution of $X$ given $Y=y$ is a multivariate normal distribution $\mathcal{N}(\mu^*, \Sigma_{X|Y})$:   Conditional Mean ($\mu^*$):$$\mu^* = \mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(y - \mu_Y)$$
This represents the best linear predictor of $X$ given $Y$.  Conditional Covariance ($\Sigma_{X|Y}$):$$\Sigma_{X|Y} = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$$
This is the Schur complement of $\Sigma_{YY}$ in $\Sigma$. Notably, the conditional covariance does not depend on the specific value of $y$.  Final Conditional Density FunctionSubstituting these parameters back into the multivariate normal form, we obtain:$$f_{X|Y}(x|y) = \frac{1}{\sqrt{(2\pi)^{n}|\Sigma_{X|Y}|}} \exp\left(-\frac{1}{2}(x - \mu^*)^{T} \Sigma_{X|Y}^{-1} (x - \mu^*)\right)$$

##Q3) Continuous Fault Monitoring of a Spindle Motor
###1. Joint Distribution
Identify the distribution of $(X, Y)$: Since $X \sim \mathcal{N}(0, 1)$ and $Q \sim \mathcal{N}(0, 4)$ are independent normal random variables, their joint distribution $(X, Y)$ is a Multivariate Normal Distribution because any linear combination of $X$ and $Y = X + Q$ is also normally distributed.  Show that $Y \sim \mathcal{N}(0, 5)$:Mean: $E[Y] = E[X + Q] = E[X] + E[Q] = 0 + 0 = 0$.  Variance: Since $X$ and $Q$ are independent, $Var(Y) = Var(X) + Var(Q) = 1 + 4 = 5$.  Therefore, $Y \sim \mathcal{N}(0, 5)$.  Find the covariance $Cov(X, Y)$:$Cov(X, Y) = Cov(X, X + Q) = Cov(X, X) + Cov(X, Q)$.  Since $X$ and $Q$ are independent, $Cov(X, Q) = 0$.  $Cov(X, Y) = Var(X) = 1$

###2. Information Geometry  
  Why $X$ is not $\sigma(Y)$-measurable: For $X$ to be $\sigma(Y)$-measurable, it would need to be a deterministic function of $Y$. However, the measurement $Y$ is "noisy" due to the independent mechanical load $Q$.  Practical Terms: If $Y = 5$, this total power could be produced by many different combinations of temperature ($X$) and load ($Q$). For example, the motor could be very hot ($X=4$) with a light load ($Q=1$), or at normal temperature ($X=0$) with a very heavy load ($Q=5$). Because different outcomes $\omega$ produce the same $Y$ but different $X$, the engineer cannot distinguish the source of the high power reading with certainty

##3. Conditional Expectation (The Best Estimate)
Compute $E[X|Y]$: Using the formula $E[X|Y=y] = \mu_X + \frac{Cov(X, Y)}{Var(Y)}(y - \mu_Y)$:  $E[X|Y=y] = 0 + \frac{1}{5}(y - 0) = \frac{1}{5}y$.  Signal-to-total-variance ratio:$E[X|Y] = \left( \frac{Var(X)}{Var(X) + Var(Q)} \right) Y = \frac{1}{1 + 4} Y = 0.2Y$.  This shows that the best estimate "shrinks" the observed measurement by 80% because most of the variance in $Y$ is assumed to come from the noisy load $Q$

##4. Verification of the Residual
Prove $E[Z] = 0$:$E[Z] = E[X - \hat{X}] = E[X] - E[E[X|Y]]$.  By the Law of Iterated Expectations, $E[E[X|Y]] = E[X] = 0$.Thus, $E[Z] = 0 - 0 = 0$.
###Verify the Orthogonality Principle
($Cov(Z, Y) = 0$):$Cov(Z, Y) = Cov(X - 0.2Y, Y) = Cov(X, Y) - 0.2Var(Y)$.  $Cov(Z, Y) = 1 - 0.2(5) = 1 - 1 = 0$.  The estimation error is uncorrelated with the observation.

##5. Information Theory: Differential Entropy
Compute $h(X)$: For a normal distribution $\mathcal{N}(\mu, \sigma^2)$, $h(X) = \frac{1}{2} \ln(2\pi e \sigma^2)$.$h(X) = \frac{1}{2} \ln(2\pi e (1)) \approx 1.4189$ nats.  Compute $h(X|Y)$:
## find the conditional variance
$\sigma_{X|Y}^2 = \sigma_X^2 - \frac{Cov(X, Y)^2}{Var(Y)}$:  $\sigma_{X|Y}^2 = 1 - \frac{1^2}{5} = 1 - 0.2 = 0.8$.$h(X|Y) = \frac{1}{2} \ln(2\pi e (0.8)) \approx 1.3073$ nats

##6. Mutual Information
#####$I(X; Y)$: $I(X; Y) = h(X) - h(X|Y) = 1.4189 - 1.3073 = 0.1116$ nats.
Percentage of uncertainty removed: $\frac{I(X; Y)}{h(X)} = \frac{0.1116}{1.4189} \approx 7.87\%$.
####Change if $Var(Q) = 100$: The conditional variance would become $\sigma_{X|Y}^2 = 1 - \frac{1^2}{101} \approx 0.99$. The uncertainty removed would drop significantly (to $\approx 0.5\%$) because the measurement $Y$ becomes dominated by noise.

##7. Diagnostic InterpretationBiased Estimator:
###$Y$ is a biased estimator
 of $X$ because $E[Y|X] = E[X + Q | X] = X + E[Q] = X$, but we want the estimate of $X$ given $Y$. Simply using $Y$ as an estimate of $X$ would consistently overestimate the magnitude of $X$ because it fails to account for the noise $Q$.
###Shrinkage:
  $E[X|Y]$ "shrinks" the measurement toward zero (the mean) because the estimator accounts for the fact that a high $Y$ is statistically more likely to be caused by a large noise $Q$ than by an extreme temperature $X$, given their relative variances.
