## Q1) The Probability of Shooting at a Target

### 1. Uniform Distribution by Area
Total Area = $\pi R^2 = \pi(10^2) = 100\pi$
Area within 1 unit of boundary = $\pi(10^2) - \pi(9^2) = 19\pi$
$P(A) = \frac{19\pi}{100\pi} = 0.19$

### 2. Uniform Distribution by Distance
The distance $r$ is chosen from $[0, 10]$. The event $A$ is $9 \le r \le 10$.
$P(A) = \frac{10-9}{10-0} = \frac{1}{10} = 0.10$

### 3. Why they differ
The answers differ because the Area model gives more "weight" to the outer edges (where there is more physical space), while the Distance model treats the center and the edge exactly the same.


# Q2) Bertrand's Paradox

The paradox demonstrates that "at random" is ill-defined without a specific coordinate system or selection mechanism. We compare the chord length to the side of an inscribed equilateral triangle.

* **Method 1 (Random Endpoints):** Choose two points on the circumference.
    * **Logic:** Fix one point. The chord is longer than the triangle side only if the second point falls within the middle $1/3$ of the circumference.
    * **Probability:** **1/3**
    

* **Method 2 (Random Radius):** Choose a radius and a point on it to be the midpoint.
    * **Logic:** The chord is longer than the side if the midpoint is chosen within the inner half of the radius (from the center to $R/2$).
    * **Probability:** **1/2**
    

* **Method 3 (Random Midpoint):** Choose a point anywhere inside the disk to be the midpoint.
    * **Logic:** The chord exceeds the side length only if the midpoint falls within an inner circle of radius $R/2$. Since area depends on the square of the radius, this inner area is $(1/2)^2 = 1/4$ of the total area.
    * **Probability:** **1/4**
    

**Conclusion:** Each model is mathematically "correct" but represents a different physical way of choosing a chord. It proves that a "uniform" distribution depends entirely on which parameter you define as uniform.

# Q8) Probability of Being a Girl?

**Problem:** A family has two children. What is the probability that both are girls, given that **at least one** of them is a girl?

**Sample Space ($S$):** The possible combinations of two children are $\{BB, BG, GB, GG\}$. Each is equally likely ($1/4$).

**The Condition:** We are given that **at least one** is a girl. This excludes the $\{BB\}$ case. Our reduced sample space is:
$$\text{Conditioned Space} = \{BG, GB, GG\}$$
**The Event:** We want the probability of $\{GG\}$.
**Calculation:**
There is 1 favorable outcome out of 3 possible outcomes in the conditioned space.
$$P(\text{Both Girls} | \text{At Least One Girl}) = \frac{1}{3}$$

# Q9) Information Available in a Discrete Random Variable

**Given:** $\Omega = \{G_1, G_2, D_1, D_2\}$ with probabilities:
$P(G_1)=0.50, P(G_2)=0.20, P(D_1)=0.10, P(D_2)=0.20$.

### 1. Probability Space $(\Omega, \mathcal{F}, P)$
* **Sample Space:** $\Omega = \{G_1, G_2, D_1, D_2\}$
* **Sigma-algebra:** $\mathcal{F} = \mathcal{P}(\Omega)$ (Power set containing 16 subsets).
* **Probability Measure:** Defined by the point masses provided above.

### 2. $\sigma$-algebra generated by $X$, $\sigma(X)$
$X$ groups outcomes into {Good} and {Defective}.
$\sigma(X) = \{\emptyset, \{G_1, G_2\}, \{D_1, D_2\}, \Omega\}$

### 3. $\sigma$-algebra generated by $Y$, $\sigma(Y)$
$Y$ distinguishes every single outcome.
$\sigma(Y) = \mathcal{P}(\Omega)$ (All 16 possible subsets of $\Omega$).

### 4. Show $\sigma(X) \subseteq \sigma(Y)$
Every set in $\sigma(X)$ is also in $\sigma(Y)$. $Y$ contains more information because it identifies the specific machine, whereas $X$ only identifies quality. Knowing $Y$ tells you $X$, but knowing $X$ does not tell you $Y$.

### 5. Marginal Distribution of $X$
* $P(X=0) = P(G_1) + P(G_2) = 0.50 + 0.20 = \mathbf{0.70}$
* $P(X=1) = P(D_1) + P(D_2) = 0.10 + 0.20 = \mathbf{0.30}$

### 6. Marginal Distribution of $Y$
* $P(Y=1)=0.50, P(Y=2)=0.20, P(Y=3)=0.10, P(Y=4)=0.20$.

### 7. Joint Probabilities
* $P(X=0, Y=1) = 0.50$
* $P(X=0, Y=2) = 0.20$
* $P(X=1, Y=3) = 0.10$
* $P(X=1, Y=4) = 0.20$

### 8. Conditional Probabilities
* $P(X=1 | Y=3) = \frac{P(X=1, Y=3)}{P(Y=3)} = \frac{0.10}{0.10} = \mathbf{1}$
* $P(X=1 | Y=4) = \frac{0.20}{0.20} = \mathbf{1}$
* $P(Y=3 | X=1) = \frac{P(X=1, Y=3)}{P(X=1)} = \frac{0.10}{0.30} = \mathbf{1/3}$
* $P(Y=4 | X=1) = \frac{0.20}{0.30} = \mathbf{2/3}$

### 9. Difference between conditioning on $X$ and $Y$
Conditioning on $Y$ removes all uncertainty about quality (it results in probability 1). Conditioning on $X$ still leaves uncertainty about which machine produced the item.

### 10. Information comparison
* **In $\sigma(Y)$ but not $\sigma(X)$:** Information about the machine origin (e.g., the set $\{G_1\}$).
* **In $\sigma(X)$ but not $\sigma(Y)$:** None, because $\sigma(X) \subseteq \sigma(Y)$.

### 12. Shannon Entropy (in bits)
* **$H(X)$:** $-(0.7\log_2 0.7 + 0.3\log_2 0.3) \approx \mathbf{0.881}$
* **$H(Y)$:** $-(0.5\log_2 0.5 + 0.2\log_2 0.2 + 0.1\log_2 0.1 + 0.2\log_2 0.2) \approx \mathbf{1.761}$
* **$H(X,Y)$:** Since $X$ is a function of $Y$, $H(X,Y) = H(Y) \approx \mathbf{1.761}$
* **$H(X|Y)$:** $0$ (Knowing the machine leaves no uncertainty about quality).

# Q10) Information Available in Continuous Random Variables

**Given:** $T \sim N(0, 1)$ and $P \sim N(0, 4)$ are independent.
Joint density: $f_{T,P}(t,p) = \frac{1}{4\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right)$.

### 1. Probability Space $(\Omega, \mathcal{F}, P)$
* **$\Omega$:** $\mathbb{R}^2$
* **$\mathcal{F}$:** $\mathcal{B}(\mathbb{R}^2)$ (Borel $\sigma$-algebra on $\mathbb{R}^2$).
* **$P$:** Probability measure induced by the joint density $f_{T,P}(t,p)$.

### 2 & 3. $\sigma$-algebras $\sigma(X)$ and $\sigma(Y)$
* **$\sigma(X)$:** The information regarding temperature deviations. It is the collection of sets $\{B \times \mathbb{R} : B \in \mathcal{B}(\mathbb{R})\}$.
* **$\sigma(Y)$:** The full sensor information. It is the entire Borel $\sigma$-algebra $\mathcal{B}(\mathbb{R}^2)$.

### 4. Show $\sigma(X) \subseteq \sigma(Y)$
Since $X$ is a coordinate projection of $Y$ ($X = \pi_1 \circ Y$), any information available in $X$ is inherently part of the full sensor state $Y$. $Y$ contains more information because it includes pressure data ($P$) which $X$ lacks.

### 5 & 6. Marginal Densities
* **$f_X(x)$:** $N(0, 1) = \frac{1}{\sqrt{2\pi}} e^{-x^2/2}$
* **$f_Y(y)$:** The joint density $f_{T,P}(t,p)$ as defined in the prompt.

### 7. Probabilities
* $P(X \le 0) = \mathbf{0.5}$ (Symmetry of Normal distribution).
* $P(X > 1) = 1 - \Phi(1) \approx \mathbf{0.1587}$.
* $P(Y \in (-\infty, 0] \times \mathbb{R}) = P(T \le 0) = \mathbf{0.5}$.
* $P(Y \in [-1, 1] \times [-2, 2]) = P(-1 \le T \le 1) \cdot P(-2 \le P \le 2) = (0.6827) \cdot (0.6827) \approx \mathbf{0.466}$.

### 8 & 9. Conditional Distributions
* **$X$ given $Y=(t,p)$:** $X$ is completely determined. It is a Dirac delta distribution $\delta_t$.
* **$Y$ given $X=t$:** Since $T, P$ are independent, the distribution of $P$ remains $N(0, 4)$. Thus, $Y|X=t \sim (t, N(0, 4))$.

### 10. Conditioning Explanation
Conditioning on $X$ narrows the state to a specific temperature but leaves pressure uncertain. Conditioning on $Y$ removes all uncertainty as both temperature and pressure are known.

### 11. Measurable Function
$X = \pi_1 \circ Y$. Because the projection $\pi_1$ is a continuous mapping, $X$ is a Borel measurable function of $Y$.

### 13 & 14. Differential Entropy
* **$h(X)$:** $\frac{1}{2}\ln(2\pi e \sigma_T^2) = \frac{1}{2}\ln(2\pi e) \approx \mathbf{1.419}$ nats.
* **$h(Y)$:** $h(T) + h(P) = \frac{1}{2}\ln(2\pi e) + \frac{1}{2}\ln(8\pi e) = \mathbf{\ln(4\pi e)}$ nats.

### 15 & 16. Conditional Differential Entropy
* **$h(Y|X)$:** $h(P) = \frac{1}{2}\ln(8\pi e)$. Knowing temperature doesn't reduce uncertainty about independent pressure.
* **$h(X|Y)$:** In the continuous case, if $X$ is determined by $Y$, $h(X|Y) \to -\infty$. This differs from the discrete case (where $H(X|Y)=0$) because a perfectly known continuous point has "zero" volume, leading to infinite negative entropy.

###12 th question python based model

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Setup for Joint Density of Y = (T, P)
# T ~ N(0, 1), P ~ N(0, 4) -> Sigma_p = 2
t = np.linspace(-3, 3, 100)
p = np.linspace(-6, 6, 100)
T, P = np.meshgrid(t, p)

# Joint Density calculation
# f(t,p) = (1/(4*pi)) * exp(-t^2/2 - p^2/8)
Z = (1 / (4 * np.pi)) * np.exp(-(T**2 / 2) - (P**2 / 8))

# Create the plots
fig = plt.figure(figsize=(18, 5))

# --- Plot 1: Joint Density of Y (Heatmap/Contour) ---
ax1 = fig.add_subplot(131)
contour = ax1.contourf(T, P, Z, cmap='viridis')
fig.colorbar(contour, ax=ax1)
ax1.set_title(r'Joint Density $f_{T,P}(t,p)$ (Full Sensor State $Y$)')
ax1.set_xlabel('Temperature Deviation ($t$)')
ax1.set_ylabel('Pressure Deviation ($p$)')

# --- Plot 2: Marginal Density of X ---
ax2 = fig.add_subplot(132)
x_axis = np.linspace(-4, 4, 100)
ax2.plot(x_axis, norm.pdf(x_axis, 0, 1), color='red', lw=2)
ax2.fill_between(x_axis, norm.pdf(x_axis, 0, 1), color='red', alpha=0.2)
ax2.set_title(r'Marginal Density of $X$ (Temperature Only)')
ax2.set_xlabel('$x$')
ax2.set_ylabel('Density')
ax2.grid(True, alpha=0.3)

# --- Plot 3: Sigma-algebra Illustration (Coarseness) ---
ax3 = fig.add_subplot(133)
# Illustrating sigma(X) as vertical strips (information only about T)
for i in np.arange(-2, 3, 1):
    # Added 'r' before the string below
    ax3.axvspan(i-0.1, i+0.1, color='blue', alpha=0.3, label=r'$\sigma(X)$ sets' if i==0 else "")

# Illustrating sigma(Y) as points/fine grid (information about T and P)
# Added 'r' before the string below
ax3.scatter(T[::10, ::10], P[::10, ::10], s=5, color='black', alpha=0.5, label=r'$\sigma(Y)$ resolution')

# Added 'r' before the string below
ax3.set_title(r'Information: $\sigma(X)$ is coarser than $\sigma(Y)$')
ax3.set_xlabel('T-axis')
ax3.set_ylabel('P-axis')
ax3.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
