# Q2) Bertrand's Paradox

The paradox demonstrates that "at random" is ill-defined without a specific coordinate system.

* **Method 1 (Random Endpoints):** Choose two points on the circumference. The probability the chord is longer than the triangle side is **1/3**.
* **Method 2 (Random Radius):** Choose a radius and a point on it. The probability is **1/2**.
* **Method 3 (Random Midpoint):** Choose a point anywhere inside the disk. The probability is **1/4**.

**Conclusion:** Each model is mathematically "correct" but represents a different physical way of choosing a chord.

# Q3) Transformer Failures

In a certain electricity distribution district, the number of emergency transformer failures reported in a single day is modeled as a discrete random variable $X$, where $X \in \{0, 1, 2, 3\}$.

**Given Probabilities:**
- $P(X = 0) = 0.50$
- $P(X = 1) = 0.30$
- $P(X = 2) = 0.15$
- $P(X = 3) = 0.05$

### 1. Construct a probability space $(\Omega, \mathcal{F}, P)$ suitable for this model.
* **Sample Space ($\Omega$):** $\Omega = \{0, 1, 2, 3\}$
* **Sigma-algebra ($\mathcal{F}$):** The power set $\mathcal{P}(\Omega)$, containing all $2^4 = 16$ possible subsets.
* **Probability Measure ($P$):** Defined by the given point masses: $P(\{0\})=0.50, P(\{1\})=0.30, P(\{2\})=0.15, P(\{3\})=0.05$. Note that $\sum P = 1.0$.

### 2. Define $X$ as a random variable on this space.
$X$ is a function $X: \Omega \to \mathbb{R}$ such that $X(\omega) = \omega$ for all $\omega \in \Omega$. It maps each outcome in the sample space to its corresponding integer value.

### 3. Determine the probability mass function (PMF) of $X$.
The PMF $p_X(x)$ is:
$$p_X(x) = \begin{cases} 0.50 & \text{if } x=0 \\ 0.30 & \text{if } x=1 \\ 0.15 & \text{if } x=2 \\ 0.05 & \text{if } x=3 \\ 0 & \text{otherwise} \end{cases}$$

### 4. Compute the expectation $E[X]$.
$$E[X] = \sum x \cdot P(X=x)$$
$$E[X] = (0 \times 0.50) + (1 \times 0.30) + (2 \times 0.15) + (3 \times 0.05)$$
$$E[X] = 0 + 0.30 + 0.30 + 0.15 = \mathbf{0.75}$$

### 5. Compute the variance $Var(X)$.
First, find $E[X^2]$:
$$E[X^2] = (0^2 \times 0.50) + (1^2 \times 0.30) + (2^2 \times 0.15) + (3^2 \times 0.05)$$
$$E[X^2] = 0 + 0.30 + 0.60 + 0.45 = 1.35$$
Now, $Var(X) = E[X^2] - (E[X])^2$:
$$Var(X) = 1.35 - (0.75)^2 = 1.35 - 0.5625 = \mathbf{0.7875}$$

### 6. Find $P(X \ge 1)$ and $P(X \ge 2)$.
* $P(X \ge 1) = 1 - P(X=0) = 1 - 0.50 = \mathbf{0.50}$
* $P(X \ge 2) = P(X=2) + P(X=3) = 0.15 + 0.05 = \mathbf{0.20}$

### 7. Briefly explain the practical meaning of the expectation and variance in this setting.
* **Expectation ($0.75$):** This represents the long-term average number of transformer failures per day. If we observe the district over many months, we expect an average of 3 failures every 4 days.
* **Variance ($0.7875$):** This measures the "spread" or uncertainty of the failures. A higher variance would indicate that the number of daily failures fluctuates wildly, making it harder for the district to plan repair resources.

# Q5) Expected Project Completion Time (Triangular Distribution)

For a project where the minimum time is $a=4$, maximum time is $b=10$, and the most likely time is $c=6$:

1. **Support of the Random Variable:** The support of $X$ is the interval $[4, 10]$ hours.

2. **Probability Density Function (PDF):**
The PDF is defined piecewise:
- For $4 \le x < 6$: $f(x) = \frac{2(x-4)}{(10-4)(6-4)} = \frac{x-4}{6}$
- For $6 \le x \le 10$: $f(x) = \frac{2(10-x)}{(10-4)(10-6)} = \frac{10-x}{12}$
- Otherwise: $f(x) = 0$

3. **Verification of PDF:** The area under the triangle must be 1.
Area = $\frac{1}{2} \times \text{Base} \times \text{Height} = \frac{1}{2} \times (10-4) \times f(6)$.
Since $f(6) = \frac{6-4}{6} = \frac{1}{3}$, Area = $\frac{1}{2} \times 6 \times \frac{1}{3} = 1$. It is a valid PDF.

4. **Expected Completion Time $E[X]$:**
$$E[X] = \frac{a+b+c}{3} = \frac{4+10+6}{3} = \frac{20}{3} \approx 6.67 \text{ hours}$$

5. **Variance $Var(X)$:**
$$Var(X) = \frac{a^2+b^2+c^2 - ab - bc - ca}{18} = \frac{16+100+36 - 40 - 60 - 24}{18} = \frac{28}{18} \approx 1.56$$

6. **Physical Interpretation of $a, b, c$:**
- **$a=4$**: Optimistic time (everything goes perfectly).
- **$b=10$**: Pessimistic time (worst-case scenario/delays).
- **$c=6$**: Most likely time (peak of the distribution).

7. **Reasonableness of the Model:** The Triangular distribution is very useful in project management (like PERT) because it is easier for experts to estimate a "min," "max," and "mode" than to estimate a mean and standard deviation.I add the Beta model using python code at the end .

8. **Probability of completion within 5 hours $P(X \le 5)$:**
Using the first part of the PDF:
$$P(X \le 5) = \int_{4}^{5} \frac{x-4}{6} dx = \left[ \frac{(x-4)^2}{12} \right]_{4}^{5} = \frac{1}{12} \approx 0.0833$$

9. **Probability $P(5 \le X \le 8)$:**
First, calculate $P(X > 8)$ using the second part of the PDF:
$$P(X > 8) = \int_{8}^{10} \frac{10-x}{12} dx = \left[ -\frac{(10-x)^2}{24} \right]_{8}^{10} = \frac{4}{24} = \frac{1}{6}$$
Then, $P(X \le 8) = 1 - \frac{1}{6} = \frac{5}{6}$.
Finally, $P(5 \le X \le 8) = P(X \le 8) - P(X \le 5) = \frac{10}{12} - \frac{1}{12} = \frac{9}{12} = 0.75$.

10. **Application in Risk Management:** This model allows a manager to quantify risk. For example, knowing there is only an 8.3% chance of finishing in 5 hours helps in setting more realistic client expectations.

# Q6) Expected Project Completion Time (PERT Distribution)

In this model, we use the PERT distribution with parameters $a=8$ (optimistic), $b=20$ (pessimistic), and $c=11$ (most likely).

1. **Support of the Random Variable:** The support of $X$ is $[8, 20]$ days.

2. **Relationship to Beta Distribution:** The PERT distribution is a special case of the Beta distribution, scaled to the interval $[a, b]$.

3. **Shape Parameter $\alpha$:**
$$\alpha = 1 + 4\left(\frac{c-a}{b-a}\right) = 1 + 4\left(\frac{11-8}{20-8}\right) = 1 + 4\left(\frac{3}{12}\right) = 2$$

4. **Shape Parameter $\beta$:**
$$\beta = 1 + 4\left(\frac{b-c}{b-a}\right) = 1 + 4\left(\frac{20-11}{20-8}\right) = 1 + 4\left(\frac{9}{12}\right) = 4$$

5. **Probability Density Function (PDF):**
The PDF is defined as:
$$f(x) = \frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}{B(\alpha, \beta)(b-a)^{\alpha+\beta-1}}$$
For our values: $f(x) = \frac{(x-8)^{1}(20-x)^{3}}{B(2, 4)(12)^5}$ for $8 \le x \le 20$.

6. **Expected Completion Time $E[X]$:**
Unlike the Triangular distribution, PERT weights the mode four times more than the extremes:
$$E[X] = \frac{a+4c+b}{6} = \frac{8+44+20}{6} = \frac{72}{6} = 12 \text{ days}$$

7. **Variance $Var(X)$:**
$$Var(X) = \frac{(E[X]-a)(b-E[X])}{7} = \frac{(12-8)(20-12)}{7} = \frac{4 \times 8}{7} \approx 4.57$$

8. **Standard Deviation:**
$$\sigma = \sqrt{4.57} \approx 2.14 \text{ days}$$

9. **Comparison with Triangular:** The PERT distribution produces a "smoother" bell-like curve compared to the sharp peak of the Triangular distribution. It is generally considered more accurate for human-estimated durations.

10. **Interpretation:** An expected value of 12 days means that while the "most likely" result is 11, the long "tail" toward 20 days (pessimistic) pulls the average completion time higher.

# Q7) PERT Expectation and Variance Intuition

Suppose we are using a PERT distribution where the optimistic time is $a=5$ and the most likely time is $c=11$.

1. **Find the pessimistic time $b$ if $E[X] = 11$.**
The formula for the PERT mean is $E[X] = \frac{a + 4c + b}{6}$.
Given $E[X] = 11, a = 5,$ and $c = 11$:
$$11 = \frac{5 + 4(11) + b}{6}$$
$$66 = 5 + 44 + b$$
$$66 = 49 + b$$
$$b = 17$$
**Answer:** The pessimistic time $b$ is **17**.

2. **What happenes to the variance in this model.**
The variance for this symmetric PERT distribution is calculated as:
$$Var(X) = \frac{(E[X] - a)(b - E[X])}{7}$$
Substituting our values:
$$Var(X) = \frac{(11 - 5)(17 - 11)}{7} = \frac{6 \times 6}{7} = \frac{36}{7} \approx 5.14$$
The variance represents the uncertainty of the project duration. In this specific case, because the mean equals the mode, the distribution is symmetric, and the variance is balanced around the most likely outcome.

# Q9) Information Available in a Discrete Random Variable

**Given:** $\Omega = \{G_1, G_2, D_1, D_2\}$ with probabilities:
$P(G_1)=0.50, P(G_2)=0.20, P(D_1)=0.10, P(D_2)=0.20$.

### 1. Probability Space $(\Omega, \mathcal{F}, P)$
* **Sample Space:** $\Omega = \{G_1, G_2, D_1, D_2\}$
* **Sigma-algebra:** $\mathcal{F} = \mathcal{P}(\Omega)$ (Power set containing 16 subsets).
* **Probability Measure:** Defined by the point masses provided above.

### 2. $\sigma$-algebra generated by $X$, $\sigma(X)$
$X$ groups outcomes into {Good} and {Defective}.
$\sigma(X) = \{\emptyset, \{G_1, G_2\}, \{D_1, D_2\}, \Omega\}$

### 3. $\sigma$-algebra generated by $Y$, $\sigma(Y)$
$Y$ distinguishes every single outcome.
$\sigma(Y) = \mathcal{P}(\Omega)$ (All 16 possible subsets of $\Omega$).

### 4. Show $\sigma(X) \subseteq \sigma(Y)$
Every set in $\sigma(X)$ is also in $\sigma(Y)$. $Y$ contains more information because it identifies the specific machine, whereas $X$ only identifies quality. Knowing $Y$ tells you $X$, but knowing $X$ does not tell you $Y$.

### 5. Marginal Distribution of $X$
* $P(X=0) = P(G_1) + P(G_2) = 0.50 + 0.20 = \mathbf{0.70}$
* $P(X=1) = P(D_1) + P(D_2) = 0.10 + 0.20 = \mathbf{0.30}$

### 6. Marginal Distribution of $Y$
* $P(Y=1)=0.50, P(Y=2)=0.20, P(Y=3)=0.10, P(Y=4)=0.20$.

### 7. Joint Probabilities
* $P(X=0, Y=1) = 0.50$
* $P(X=0, Y=2) = 0.20$
* $P(X=1, Y=3) = 0.10$
* $P(X=1, Y=4) = 0.20$

### 8. Conditional Probabilities
* $P(X=1 | Y=3) = \frac{P(X=1, Y=3)}{P(Y=3)} = \frac{0.10}{0.10} = \mathbf{1}$
* $P(X=1 | Y=4) = \frac{0.20}{0.20} = \mathbf{1}$
* $P(Y=3 | X=1) = \frac{P(X=1, Y=3)}{P(X=1)} = \frac{0.10}{0.30} = \mathbf{1/3}$
* $P(Y=4 | X=1) = \frac{0.20}{0.30} = \mathbf{2/3}$

### 9. Difference between conditioning on $X$ and $Y$
Conditioning on $Y$ removes all uncertainty about quality (it results in probability 1). Conditioning on $X$ still leaves uncertainty about which machine produced the item.

### 10. Information comparison
* **In $\sigma(Y)$ but not $\sigma(X)$:** Information about the machine origin (e.g., the set $\{G_1\}$).
* **In $\sigma(X)$ but not $\sigma(Y)$:** None, because $\sigma(X) \subseteq \sigma(Y)$.

### 12. Shannon Entropy (in bits)
* **$H(X)$:** $-(0.7\log_2 0.7 + 0.3\log_2 0.3) \approx \mathbf{0.881}$
* **$H(Y)$:** $-(0.5\log_2 0.5 + 0.2\log_2 0.2 + 0.1\log_2 0.1 + 0.2\log_2 0.2) \approx \mathbf{1.761}$
* **$H(X,Y)$:** Since $X$ is a function of $Y$, $H(X,Y) = H(Y) \approx \mathbf{1.761}$
* **$H(X|Y)$:** $0$ (Knowing the machine leaves no uncertainty about quality).

# Q10) Information Available in Continuous Random Variables

**Given:** $T \sim N(0, 1)$ and $P \sim N(0, 4)$ are independent.
Joint density: $f_{T,P}(t,p) = \frac{1}{4\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right)$.

### 1. Probability Space $(\Omega, \mathcal{F}, P)$
* **$\Omega$:** $\mathbb{R}^2$
* **$\mathcal{F}$:** $\mathcal{B}(\mathbb{R}^2)$ (Borel $\sigma$-algebra on $\mathbb{R}^2$).
* **$P$:** Probability measure induced by the joint density $f_{T,P}(t,p)$.

### 2 & 3. $\sigma$-algebras $\sigma(X)$ and $\sigma(Y)$
* **$\sigma(X)$:** The information regarding temperature deviations. It is the collection of sets $\{B \times \mathbb{R} : B \in \mathcal{B}(\mathbb{R})\}$.
* **$\sigma(Y)$:** The full sensor information. It is the entire Borel $\sigma$-algebra $\mathcal{B}(\mathbb{R}^2)$.

### 4. Show $\sigma(X) \subseteq \sigma(Y)$
Since $X$ is a coordinate projection of $Y$ ($X = \pi_1 \circ Y$), any information available in $X$ is inherently part of the full sensor state $Y$. $Y$ contains more information because it includes pressure data ($P$) which $X$ lacks.

### 5 & 6. Marginal Densities
* **$f_X(x)$:** $N(0, 1) = \frac{1}{\sqrt{2\pi}} e^{-x^2/2}$
* **$f_Y(y)$:** The joint density $f_{T,P}(t,p)$ as defined in the prompt.

### 7. Probabilities
* $P(X \le 0) = \mathbf{0.5}$ (Symmetry of Normal distribution).
* $P(X > 1) = 1 - \Phi(1) \approx \mathbf{0.1587}$.
* $P(Y \in (-\infty, 0] \times \mathbb{R}) = P(T \le 0) = \mathbf{0.5}$.
* $P(Y \in [-1, 1] \times [-2, 2]) = P(-1 \le T \le 1) \cdot P(-2 \le P \le 2) = (0.6827) \cdot (0.6827) \approx \mathbf{0.466}$.

### 8 & 9. Conditional Distributions
* **$X$ given $Y=(t,p)$:** $X$ is completely determined. It is a Dirac delta distribution $\delta_t$.
* **$Y$ given $X=t$:** Since $T, P$ are independent, the distribution of $P$ remains $N(0, 4)$. Thus, $Y|X=t \sim (t, N(0, 4))$.

### 10. Conditioning Explanation
Conditioning on $X$ narrows the state to a specific temperature but leaves pressure uncertain. Conditioning on $Y$ removes all uncertainty as both temperature and pressure are known.

### 11. Measurable Function
$X = \pi_1 \circ Y$. Because the projection $\pi_1$ is a continuous mapping, $X$ is a Borel measurable function of $Y$.

### 13 & 14. Differential Entropy
* **$h(X)$:** $\frac{1}{2}\ln(2\pi e \sigma_T^2) = \frac{1}{2}\ln(2\pi e) \approx \mathbf{1.419}$ nats.
* **$h(Y)$:** $h(T) + h(P) = \frac{1}{2}\ln(2\pi e) + \frac{1}{2}\ln(8\pi e) = \mathbf{\ln(4\pi e)}$ nats.

### 15 & 16. Conditional Differential Entropy
* **$h(Y|X)$:** $h(P) = \frac{1}{2}\ln(8\pi e)$. Knowing temperature doesn't reduce uncertainty about independent pressure.
* **$h(X|Y)$:** In the continuous case, if $X$ is determined by $Y$, $h(X|Y) \to -\infty$. This differs from the discrete case (where $H(X|Y)=0$) because a perfectly known continuous point has "zero" volume, leading to infinite negative entropy.

# Q11) Conditional Expectation

**Given:** $\Omega = \{\omega_1, \omega_2, \omega_3, \omega_4\}$ with probabilities:
$P(\omega_1)=0.50, P(\omega_2)=0.10, P(\omega_3)=0.20, P(\omega_4)=0.20$.

* **$X$ (Fraud):** $X(\omega_1)=0, X(\omega_2)=1, X(\omega_3)=0, X(\omega_4)=1$.
* **$Y$ (Risk Class):** $Y(\omega_1)=0, Y(\omega_2)=0, Y(\omega_3)=1, Y(\omega_4)=1$.

### 1. Probability Space $(\Omega, \mathcal{F}, P)$
* **Sample Space:** $\Omega = \{\omega_1, \omega_2, \omega_3, \omega_4\}$
* **Sigma-algebra:** $\mathcal{F} = \mathcal{P}(\Omega)$ (Power set).
* **Measure:** $P$ is the discrete measure defined by the given probabilities.

### 2. $\sigma$-algebra generated by $Y$, $\sigma(Y)$
$Y$ identifies groups of outcomes: $\{\omega_1, \omega_2\}$ (Low-risk) and $\{\omega_3, \omega_4\}$ (High-risk).
$\sigma(Y) = \{ \emptyset, \{\omega_1, \omega_2\}, \{\omega_3, \omega_4\}, \Omega \}$

### 3. Measurability of $E[X|Y]$
$E[X|Y]$ is the best estimate of $X$ given $Y$. Because $Y$ cannot distinguish between $\omega_1$ and $\omega_2$, any function measurable with respect to $\sigma(Y)$ must assign the same value to both outcomes. Thus, $E[X|Y]$ is constant on the atoms of $\sigma(Y)$.

### 4. Compute $E[X|Y]$ explicitly
* **For $Y=0$ (Low-risk):**
$$E[X | Y=0] = \frac{X(\omega_1)P(\omega_1) + X(\omega_2)P(\omega_2)}{P(\omega_1) + P(\omega_2)} = \frac{0(0.5) + 1(0.1)}{0.5 + 0.1} = \frac{0.1}{0.6} = \mathbf{0.1667}$$
* **For $Y=1$ (High-risk):**
$$E[X | Y=1] = \frac{X(\omega_3)P(\omega_3) + X(\omega_4)P(\omega_4)}{P(\omega_3) + P(\omega_4)} = \frac{0(0.2) + 1(0.2)}{0.2 + 0.2} = \frac{0.2}{0.4} = \mathbf{0.50}$$

### 5. Verify the defining property
For $A = \{\omega_1, \omega_2\}$:
$\int_A E[X|Y] dP = (0.1667 \times 0.6) = 0.1$.
$\int_A X dP = X(\omega_1)P(\omega_1) + X(\omega_2)P(\omega_2) = 0 + 0.1 = 0.1$. **Verified.**

### 6. Compute $E[X]$ and compare
$E[X] = (0 \times 0.5) + (1 \times 0.1) + (0 \times 0.2) + (1 \times 0.2) = \mathbf{0.30}$.
$E[X]$ is the global average fraud rate, while $E[X|Y]$ provides the fraud rate specific to the customer's risk category.

### 7. Practical Interpretation
It tells the bank the **probability of fraud** based only on the customer's risk class. A low-risk customer has a 16.7% chance of the transaction being fraud, while a high-risk customer has a 50% chance.

### 8. Compute $E[X | \mathcal{G}]$ where $\mathcal{G} = \mathcal{P}(\Omega)$
Since $\mathcal{G}$ contains all information (the full outcome $\omega$ is known), $E[X | \mathcal{G}] = X$.
* $E[X | \mathcal{G}](\omega_1) = 0$, $(\omega_2) = 1$, $(\omega_3) = 0$, $(\omega_4) = 1$.
Compared to $E[X|Y]$, this is a "perfect" estimate because it uses full information, whereas $E[X|Y]$ is blurred by the risk-class grouping.

### 9. Best Estimate Explanation
This example shows that $E[X|Y]$ takes the actual fraud values and "averages" them within the groups defined by $Y$. It minimizes the mean squared error based on the limited information available in $\sigma(Y)$, effectively providing the most accurate "guess" possible without knowing the full state of $\omega$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta, triang, norm

# Create a figure with 3 plots
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

# 1. Beta Distribution (Q4)
x_beta = np.linspace(0, 1, 100)
ax[0].plot(x_beta, beta.pdf(x_beta, 2, 18), 'r-', lw=3)
ax[0].set_title("Q4: Beta Distribution (α=2, β=18)")
ax[0].fill_between(x_beta, beta.pdf(x_beta, 2, 18), color='red', alpha=0.2)

# 2. Triangular Distribution (Q5)
x_tri = np.linspace(4, 10, 100)
# In scipy, c is the mode as a fraction of the range
ax[1].plot(x_tri, triang.pdf(x_tri, (6-4)/(10-4), loc=4, scale=6), 'g-', lw=3)
ax[1].set_title("Q5: Triangular Distribution (4, 6, 10)")
ax[1].fill_between(x_tri, triang.pdf(x_tri, (6-4)/(10-4), loc=4, scale=6), color='green', alpha=0.2)

# 3. Normal Distribution (Q10)
x_norm = np.linspace(-4, 4, 100)
ax[2].plot(x_norm, norm.pdf(x_norm, 0, 1), 'b-', lw=3)
ax[2].set_title("Q10: Normal Distribution (μ=0, σ=1)")
ax[2].fill_between(x_norm, norm.pdf(x_norm, 0, 1), color='blue', alpha=0.2)

plt.tight_layout()
plt.show()
```
